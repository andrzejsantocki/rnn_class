```python
import ipywidgets as widgets
from IPython.display import display
```

```python
uploader = widgets.FileUpload(
    accept='.txt',
    multiple=False
)

text_area = widgets.Textarea(
    value='Upload a .txt file to see its content here.',
    placeholder='File content will be displayed here.',
    description='',
    disabled=False,
    layout=widgets.Layout(width='100%', height='400px', border='solid 1px #cccccc')
)

def on_upload_change(change):
    if not change['new']:
        return
    uploaded_file_info = next(iter(uploader.value.values()))
    content = uploaded_file_info['content'].decode('utf-8')
    text_area.value = content
    text_area.layout.border = 'solid 1px #4F4F4F'
    text_area.layout.background_color = '#2E2E2E'
    text_area.style.text_color = '#E0E0E0'

uploader.observe(on_upload_change, names='value')

display(uploader, text_area)
```
